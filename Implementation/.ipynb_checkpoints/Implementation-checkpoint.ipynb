```python
import time

from random import sample, choices, shuffle
from itertools import permutations, product
from numpy import array
from sympy.utilities.iterables import multiset_permutations
from scipy.special import comb
from IPython.display import clear_output
import gc

from Sampling import GFsample, create_vec, create_vec_d, inv_create_vec
from Searchspace_representations import *
from Multinomial import multinomial
from Eps import leveleps, nulleps
from Cbd_stuff import cbdtuple
from Nw_print import Nw_print
from Sort_search import *
from Lists import *
from Merge import *

def vecsample(q,m):
    out=[]
    for i in range(m):
        out+=[k(GFsample(q))]
    return array(out)

def matsample(q,m,n):
    out=[]
    for i in range(n):
        out+=[vecsample(q,m)]
    return array(out)

def Gen(N, eta, q, rf_1, pkonly=False, half=False):
    A=matsample(q,N,N)
    e=array(cbdtuple(eta,N-rf[1],q)+rf[1]*[(0)%q])
    if half:
        s=array(create_vec(scale_list(Nw[0],1/2), q, shuffling=True)+create_vec(scale_list(Nw[0],1/2), q, shuffling=True))
    else:
        s=array(create_vec(scale_list(Nw[0],1), q, shuffling=True))
    t=A@s+e
    pk=A,t
    sk=s,e
    if pkonly:
        return pk
    else:
        return pk, sk
```

```python
def meet(
    A,
    t,
    d,
    rf,
    Nw,
    max_iter=100000,
    half=False
):
    global g, tq, avg_L, L, full_out
    full_out={}
    N=len(t)
    nrf=[N-rf[i] for i in range(len(rf))]
    rrf=[0,0]+[rf[i-1]-rf[i] for i in range(2,len(rf))]
    iteration_counter=0
    out=[[]]
    while iteration_counter<max_iter and out[0]==[]:
        del(out)
        gc.collect()
        clear_output(wait=True)
        iteration_counter+=1
        print("Iteration " + str(iteration_counter))
        eprime=array(rf[1]*[0]) #denotes the guessed rf[1] many entries of e. For us, this is the zero vector.

        #randomly sample collision vectors
        tq={}
        for j in range(1,d):
            for i in range(1,2**(j-1)+1):
                tq[i,j]=vecsample(q,rf[j])

        #create goal vectors for each list
        g={}
        g[1,1]=tq[1,1]
        g[2,1]=t[nrf[1]:]-eprime-tq[1,1]
        for j in range(2,d):
            for i in range(1,2**(j-1)+1):
                g[2*i-1,j]=tq[i,j]
                g[2*i,j]=g[i,j-1][rrf[j]:]-tq[i,j]

        #create base lists
        L={}
        Llen={}
        L[1,d]={key:[] for key in range(q**rf[d-1])}
        for i in range(1,2**(d-1)+1):
            L[2*i,d]={key:[] for key in range(q**rf[d-1])}
#        veccount=0
        if half==True:
            for vec in multiset_permutations(create_vec_d(Nw[d], q)):
#                veccount+=1
                v=array(vec+N//2*[0])
                Av=A@v
                c=assign_int(Av[nrf[d-1]:],q)
                L[1,d][c]+=[(v, Av, assign_int(vec,q))]
                v=array(N//2*[0]+vec)
                Av=A@v
                for i in range(1,2**(d-1)+1):
                    c=assign_int(g[i,d-1]-(Av)[nrf[d-1]:],q)
                    L[2*i,d][c]+=[(v, Av, assign_int(vec,q))]
#                if veccount%100==0:
#                    clear_output(wait=True)
#                    print(iteration_counter)
#                    print(veccount)
            for key in L[1,d]:
                L[1,d][key].sort(key = lambda x: x[2])
        else:
            for vec in multiset_permutations(create_vec(Nw[d],q)):
#                veccount+=1
                v=array(vec)
                Av=A@v
                c=assign_int(Av[nrf[d-1]:],q)
                L[1,d][c]+=[(v, Av, assign_int(vec,q))]
                for i in range(1,2**(d-1)+1):
                    c=assign_int(g[i,d-1]-(Av)[nrf[d-1]:],q)
                    L[2*i,d][c]+=[(v, Av, assign_int(vec,q))]
#                if veccount%100==0:
#                    clear_output(wait=True)
#                    print(iteration_counter)
#                    print(veccount)
            for key in L[1,d]:
                L[1,d][key].sort(key = lambda x: x[2])
        print("---------- " + str(d) + " ----------")
        for i in range(1,2**(d-1)+1):
            L[2*i-1,d]=L[1,d]
            Llen[2*i-1,d]=diclen(L[2*i-1,d])
            for key in L[2*i,d]:
                L[2*i,d][key].sort(key = lambda x: x[2])
            Llen[2*i,d]=diclen(L[2*i,d])
        print(Llen[1,d])
        avg_L[d]+=[Llen[i,d] for i in range(1,2**d+1)]
        
        
        # merge lists until 4 level 2 lists result
        for negj in range(1, d-1):
            j=d-negj
            print("---------- " + str(j) + " ----------")
            for i in range(1,2**(j-1)+1):
                true_i=2*i-1
                L[true_i,j]=merge_lists(L[2*true_i-1,j+1],L[2*true_i,j+1],nrf[j-1], Nw[j], 0, rf[j-1]*[0], q)
                Llen[true_i,j]=diclen(L[true_i,j])
                print(Llen[true_i,j])
                avg_L[j]+=[Llen[true_i,j]]
                true_i=2*i
                L[true_i,j]=merge_lists(L[2*true_i-1,j+1],L[2*true_i,j+1],nrf[j-1], Nw[j], 1, g[i, j-1], q)
                Llen[true_i,j]=diclen(L[true_i,j])
                print(Llen[true_i,j])
                avg_L[j]+=[Llen[true_i,j]]
            print("")
            print(float(avglen([Llen[i,j] for i in range(1,2**j+1)])))
            print("")
            print("")
                
        print("---------- " + str(1) + " ----------")
        L[1,1]=merge_lists_level_2(L[1,2],L[2,2], Nw[1], 0, N*[0], q, eta)
        Llen[1,1]=diclen(L[1,1])
        print(Llen[1,1])
        avg_L[1]+=[Llen[1,1]]
        L[2,1]=merge_lists_level_2(L[3,2],L[4,2], Nw[1], 1, t, q, eta)
        Llen[2,1]=diclen(L[2,1])
        print(Llen[2,1])
        avg_L[1]+=[Llen[2,1]]
        print("")
        print(float(avglen([Llen[1,1],Llen[2,1]])))
        print("")
        print("")
        print("---------- " + str(0) + " ----------")
        out=merge_lists_level_1(L[1,1],L[2,1], Nw[0], t, q, eta)
        avg_L[0]+=[len(out[0])]
        print("")
        print(len(out[0]), len(out[1]))
        print("")
        print("")
        full_out[iteration_counter]=out
    return full_out
```
