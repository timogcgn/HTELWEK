```python
import time

from random import sample, choices, shuffle
from itertools import permutations, product
from numpy import array
from sympy.utilities.iterables import multiset_permutations
from scipy.special import comb
from IPython.display import clear_output
import gc

from Sort_search import insertion_bsearch, bsearch_component, assign_int, inv_assign_int
from Sampling import GFsample, create_vec, create_vec_d, inv_create_vec
from Cbd_stuff import centered_binom_dist, cbdsample, cbdtuple
from Uniform_stuff import unidist, unisample, unituple
from Odlyzko import odlyzko, eta_small
# from Bf_list import brute_force_list
from Multinomial import multinomial
from Eps import leveleps, nulleps
from Nw_print import Nw_print
from Lists import *

# from Params_cbd import eta, N, Nw, R, d, eps, q
# from Params_cbd import rfloor as rf

from Params_cbd_small import eta, N, Nw, R, d, eps, q
from Params_cbd_small import rfloor as rf

def vecsample(q,m):
    out=[]
    for i in range(m):
        out+=[k(GFsample(q))]
    return array(out)

def matsample(q,m,n):
    out=[]
    for i in range(n):
        out+=[vecsample(q,m)]
    return array(out)

def Gen(N, rf_1, pkonly=False):
    global q
    A=matsample(q,N,N)
    e=array(cbdtuple(eta,N-rf[1],q)+rf[1]*[(0)%q])
    # s=array(create_vec(scale_list(Nw[0],1/2), q, shuffling=True)+create_vec(scale_list(Nw[0],1/2), q, shuffling=True))
    s=array(create_vec(scale_list(Nw[0],1), q, shuffling=True))
    t=A@s+e
    pk=A,t
    sk=s,e
    if pkonly:
        return pk
    else:
        return pk, sk

k=GF(q)

avg_L=[[] for i in range(d+1)]
```

```python
def eta_small(v, eta, q):
    smallset=[(i)%q for i in range(-eta,eta+1)]
    for vi in v:
        if not(vi in smallset):
            return False
    return True
```

```python
def avglen(a):
    if len(a)>0:
        return sum(a) / len(a)
    else:
        return 0

def merge_lists(L1, L2, rf_x, nrf_x, Nw_prevlevel, sign, g):
    global q
    out={}
#    count=0
    for key in L1:
#        count+=1
#        clear_output(wait=True)
#        print(count)
        if key in L2:
            for vec1 in L1[key]:
                for vec2 in L2[key]:
                    v=vec1[0]+vec2[0]
                    Av=vec1[1]+vec2[1]
                    if(
                        weight(v,1)==weight(v,q-1) and
                        weight(v,2)==weight(v,q-2) and
                        weight(v,3)==weight(v,q-3) and
                        inv_create_vec(v)==Nw_prevlevel
                    ):
                        assint=assign_int(g+(-1)**sign*(Av)[nrf_x:],q)
                        if assint in out:
                            out[assint]=insertion_bsearch(out[assint],(v, Av, assign_int(v,q)),2)
                        else:
                            out[assint]=[(v, Av, assign_int(v,q))]

    del(L1)
    del(L2)
    gc.collect()
    return out

def merge_lists_level_2(L1, L2, Nw_prevlevel, sign, g):
    global q, eta
    out={}
    count=0
    for key in L1:
        if key in L2:
            for vec1 in L1[key]:
                for vec2 in L2[key]:
                    v=vec1[0]+vec2[0]
                    Av=vec1[1]+vec2[1]
                    if(
                        weight(v,1)==weight(v,q-1) and
                        weight(v,2)==weight(v,q-2) and
                        weight(v,3)==weight(v,q-3) and
                        inv_create_vec(v)==Nw_prevlevel
                    ):
                        count+=1
                        odlyzko_hash=odlyzko(g+(-1)**sign*Av, eta, q)
                        for assint in odlyzko_hash:
                            if assint in out:
                                out[assint]=insertion_bsearch(out[assint],(v, Av, assign_int(v,q)),2)
                            else:
                                out[assint]=[(v, Av, assign_int(v,q))]
    del(L1)
    del(L2)
    gc.collect()
    return out

def merge_lists_level_1(L1, L2, Nw_prevlevel, t):
    global q, eta
    out_yes=[]
    out_no=[]
    for key in L1:
        if key in L2:
            for vec1 in L1[key]:
                for vec2 in L2[key]:
                    v=vec1[0]+vec2[0]
                    Av=vec1[1]+vec2[1]
                    if(
                        weight(v,1)==weight(v,q-1) and
                        weight(v,2)==weight(v,q-2) and
                        weight(v,3)==weight(v,q-3) and
                        inv_create_vec(v)==Nw_prevlevel and
                        eta_small(Av-t, eta, q)
                    ):
                        out_yes+=[v]
                    else:
                        out_no+=[v]
    del(L1)
    del(L2)
    gc.collect()
    return out_yes, out_no

def Meet(
    A,
    t,
    d,
    rf,
    Nw,
    max_iter=100000,
    half=False
):
    global g, tq, avg_L, L, full_out
    full_out={}
    N=len(t)
    nrf=[N-rf[i] for i in range(len(rf))]
    rrf=[0,0]+[rf[i-1]-rf[i] for i in range(2,len(rf))]
    iteration_counter=0
    out=[[]]
    while iteration_counter<max_iter and out[0]==[]:
        del(out)
        gc.collect()
#        clear_output(wait=True)
        iteration_counter+=1
        print("Iteration " + str(iteration_counter))
        eprime=array(rf[1]*[0]) #denotes the guessed rf[1] many entries of e. For us, this is the zero vector.

        #randomly sample collision vectors
        tq={}
        for j in range(1,d):
            for i in range(1,2**(j-1)+1):
                tq[i,j]=vecsample(q,rf[j])

        #create goal vectors for each list
        g={}
        g[1,1]=tq[1,1]
        g[2,1]=t[nrf[1]:]-eprime-tq[1,1]
        for j in range(2,d):
            for i in range(1,2**(j-1)+1):
                g[2*i-1,j]=tq[i,j]
                g[2*i,j]=g[i,j-1][rrf[j]:]-tq[i,j]

        #create base lists
        L={}
        Llen={}
        L[1,d]={key:[] for key in range(q**rf[d-1])}
        for i in range(1,2**(d-1)+1):
            L[2*i,d]={key:[] for key in range(q**rf[d-1])}
#        veccount=0
        if half==True:
            for vec in multiset_permutations(create_vec_d(Nw[d])):
#                veccount+=1
                v=array(vec+N//2*[0])
                Av=A@v
                c=assign_int(Av[nrf[d-1]:],q)
                L[1,d][c]+=[(v, Av, assign_int(vec,q))]
                v=array(N//2*[0]+vec)
                Av=A@v
                for i in range(1,2**(d-1)+1):
                    c=assign_int(g[i,d-1]-(Av)[nrf[d-1]:],q)
                    L[2*i,d][c]+=[(v, Av, assign_int(vec,q))]
#                if veccount%100==0:
#                    clear_output(wait=True)
#                    print(iteration_counter)
#                    print(veccount)
            for key in L[1,d]:
                L[1,d][key].sort(key = lambda x: x[2])
        else:
            for vec in multiset_permutations(create_vec(Nw[d],q)):
#                veccount+=1
                v=array(vec)
                Av=A@v
                c=assign_int(Av[nrf[d-1]:],q)
                L[1,d][c]+=[(v, Av, assign_int(vec,q))]
                for i in range(1,2**(d-1)+1):
                    c=assign_int(g[i,d-1]-(Av)[nrf[d-1]:],q)
                    L[2*i,d][c]+=[(v, Av, assign_int(vec,q))]
#                if veccount%100==0:
#                    clear_output(wait=True)
#                    print(iteration_counter)
#                    print(veccount)
            for key in L[1,d]:
                L[1,d][key].sort(key = lambda x: x[2])
        print("---------- " + str(d) + " ----------")
        for i in range(1,2**(d-1)+1):
            L[2*i-1,d]=L[1,d]
            Llen[2*i-1,d]=diclen(L[2*i-1,d])
            for key in L[2*i,d]:
                L[2*i,d][key].sort(key = lambda x: x[2])
            Llen[2*i,d]=diclen(L[2*i,d])
        print(Llen[1,d])
        avg_L[d]+=[avglen([Llen[i,d] for i in range(1,2**d+1)])]
        # merge lists until 4 level 2 lists result
        for negj in range(1, d-1):
            j=d-negj
            print("---------- " + str(j) + " ----------")
            for i in range(1,2**(j-1)+1):
                true_i=2*i-1
                L[true_i,j]=merge_lists(L[2*true_i-1,j+1],L[2*true_i,j+1],rf[j-1],nrf[j-1], Nw[j], 0, rf[j-1]*[0])
                Llen[true_i,j]=diclen(L[true_i,j])
                print(Llen[true_i,j])
                true_i=2*i
                L[true_i,j]=merge_lists(L[2*true_i-1,j+1],L[2*true_i,j+1],rf[j-1],nrf[j-1], Nw[j], 1, g[i, j-1])
                Llen[true_i,j]=diclen(L[true_i,j])
                print(Llen[true_i,j])
            print("")
            avg_L[j]+=[float(avglen([Llen[i,j] for i in range(1,2**j+1)]))]
            print(float(avglen([Llen[i,j] for i in range(1,2**j+1)])))
            print("")
            print("")
                
        print("---------- " + str(1) + " ----------")
        L[1,1]=merge_lists_level_2(L[1,2],L[2,2], Nw[1], 0, N*[0])
        Llen[1,1]=diclen(L[1,1])
        print(Llen[1,1])
        L[2,1]=merge_lists_level_2(L[3,2],L[4,2], Nw[1], 1, t)
        Llen[2,1]=diclen(L[2,1])
        print(Llen[2,1])
        avg_L[1]+=[float(avglen([Llen[1,1],Llen[2,1]]))]
        print("")
        print(float(avglen([Llen[1,1],Llen[2,1]])))
        print("")
        print("")
        print("---------- " + str(0) + " ----------")
        out=merge_lists_level_1(L[1,1],L[2,1], Nw[0], t)
        avg_L[0]+=[len(out[0])]
        print("")
        print(len(out[0]))
        print("")
        print("")
        full_out[iteration_counter]=out
    return full_out
```

```python
def brute_force_list(
    A,
    Nw,
    g_next,
    g,
    sign,
    q
):
    N=len(create_vec(Nw,q))
    negr=N-len(g)
    negg=len(g_next)-len(g)
    negr_next=N-len(g_next)
    out={}
    counter=0
    for v in multiset_permutations(create_vec(Nw,q)):
        counter+=1
        if counter%1000==0:
            clear_output(wait=True)
            print(counter)
        v=array(v)
        Av=A@v
        if list((Av)[negr:])==list(g):
            assint=assign_int(g_next+(-1)**sign*(Av)[negr_next:],q)
            if assint in out:
                out[assint]+=[(v,Av, assign_int(v,q))]
            else:
                out[assint]=[(v,Av, assign_int(v,q))]
    return out
            
            
```
